```python
#1. Write a Python function that takes a list of numbers as input and returns the sum of all even numbers in the list.

list = [12,22,13,54,57]
def sum_even_numbers(numbers):
  s = 0
  for n in numbers:
    if n % 2 == 0:
      s = s + n
  return s
print(sum_even_numbers(list))
```

```python
#3. Implement a Python function that takes a list of integers and returns a new list containing the squares of each number.

def sq_of_list(numbers):
  squares = []
  for n in numbers:
    squares.append(n**2)
  return squares
print(sq_of_list([1,44,2,3,1]))
```

```python
#4. Write a Python function that checks if a given number is prime or not from 1 to 200.

def num_is_prime(numbers):
  if numbers >1 and numbers <201:
    for i in range(2,numbers):
      if numbers % i == 0:
        return False
    return True
```

```python
#5. Create an iterator class in Python that generates the Fibonacci sequence up to a specified number of terms.

class FibonacciIterator:
    def __init__(self, n):
        self.n = n
        self.count = 0
        self.a, self.b = 0, 1
    def __iter__(self):
        return self
    def __next__(self):
        if self.count >= self.n:
            raise StopIteration
        result = self.a
        self.a, self.b = self.b, self.a + self.b

        self.count += 1
        return result
fib = FibonacciIterator(12)
for num in fib:
    print(num)
```

```python
#6. Write a generator function in Python that yields the powers of 2 up to a given exponent.

def powers_of_2(exponent):
    for i in range(exponent + 1):
        yield 2 ** i

gen = powers_of_2(9)
for value in gen:
    print(value)
```

```python
#7. Implement a generator function that reads a file line by line and yields each line as a string.

def read_file_lines(filename):
    with open(filename, 'r') as file:
        for line in file:
            yield line
for line in read_file_lines("sample.txt"):
    print(line, end="")
```

```python
#8. Use a lambda function in Python to sort a list of tuples based on the second element of each tuple.

list = [(1, 2), (4, 1), (9, 10), (13, 12)]
sorted_list = sorted(list, key=lambda x: x[1])
print(sorted_list)
```

```python
#9. Write a Python program that uses `map()` to convert a list of temperatures from Celsius to Fahrenheit.

Celsius = [22,31,44,11]
fahrenheit =(map(lambda c: (c * 9/5) + 32, Celsius))
fahrenheit_list= __builtins__.list(fahrenheit)
print(fahrenheit)
print(fahrenheit_list)
```

```python
#10. Create a Python program that uses `filter()` to remove all the vowels from a given string.

def remove_vowels(string):
    vowels = 'aeiouAEIOU'
    filtered_string = ''.join(filter(lambda x: x not in vowels, string))
    return filtered_string
print(remove_vowels("You Never Walk Alone"))
```
